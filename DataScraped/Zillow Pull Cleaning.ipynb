```python
# Importing Pandas
import pandas as pd
```

```python
# Creating a DataFrame from the csv file
df = pd.read_csv("Houston_Properties.csv", encoding= 'unicode_escape')

df.head(20)
```

```python
# Checking data types
df.dtypes
```

```python
# Converting data types
df.convert_dtypes

df.dtypes
```

```python
# Dropping list elements with null values
df.dropna(inplace = True)

#Separating out all abberant values from a main DataFrame
buffer_df = df.loc[df['transit'] > 100]
main_df = df.loc[df['transit']<= 100]

#Normalizing values for the large number of abberants
buffer_df['walk'] = 50
buffer_df['transit'] = 50
buffer_df['bike'] = 50

#Re-merging the main and buffer DataFrames
main_df = main_df.append(buffer_df)
```

```python
main_df
```

```python
main_df.to_csv("Cleaned1.csv")
```

```python
zipcodes = list()

for item in main_df['address']:
    zipcode = item.split()[-1]
    zipcodes.append(zipcode)

main_df['zipcodes'] = zipcodes
df = pd.DataFrame(main_df)
```
